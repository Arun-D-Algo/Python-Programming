1. Write a Python function to input two numbers and perform the Calculator operations of (+, -, *, /).

```python
x = int(input("Enter first number : "))   # int(input()) used to ensure that the input (in string format) gets type casted to int
y = int(input("Enter second number : "))  # int(input()) used to ensure that the input (in string format) gets type casted to int

operator = (input("Enter operator : "))   # no type cast required here because operator is taken in as a string itself

match operator:                           # Match case used here to check for operator and accordingly operate on integers
    case '+':
        print(f"The sum of {x} and {y} is : {x+y}")
    case '-':
        print(f"The difference of {x} and {y} is : {x-y}")
    case '*':
        print(f"The product of {x} and {y} is : {x*y}")
    case '/':
        print(f"The quotient of {x} and {y} is : {x/y}")   # / in python gives the true float divison
```

2. Write a Python function that takes an integer and returns True if it’s a prime number and False otherwise.

```python
x = int(input("Enter the number : "))

def isprime(x):
    if x<2:     
        return False
    i = 2
    while i <= np.sqrt(x):          # Algorithm to find whether a number is prime or not
        if x%i == 0:
            return False
        i+=1
    return True

print("Is the number a prime : ", isprime(x))
```

3. Create a Python function that creates a sequence between 1 and 100 and prints all the odd numbers. Compute and display the sum of all the even numbers.

```python
def sequence():
    numbers = np.arange(1,101)    # np.arange used to provide the (start, stop (exclusive), step) of range of numbers to put in array
    odd = numbers[numbers%2!=0]   # since the numbers are generated in sequence, numbers%2 gives the array index as well here to print
    even = numbers[numbers%2==0]
    print(odd)
    print(np.sum(even))           # np.sum(x) gives the sum of all values in x array

sequence()
```

5. Write a Python function that takes a string input from the user and counts the number of vowels and consonants in the string.

```python
def count():
    string = input("Enter a string : ")

    vowels = 0
    consonants = 0

    for ch in string:
        if ch in "aeiouAEIOU":
            vowels += 1
        elif ch.isalpha():    # ch.isalpha() checks to make sure the character is an alphabet only
            consonants += 1

    print("The String is : ", string)
    print("The number of vowels in the string is : ", vowels)
    print("The number of consonants in the string is : ", consonants)

count()  
```
