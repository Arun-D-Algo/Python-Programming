Create a Series from a list of integers representing daily
temperatures (in Celsius) over a week. Assign index labels as day
of the week.

a. Find and print the average (mean) temperature for the week.

b. Identify and print the maximum and minimum temperatures and their respective days.

c. Display the temperatures greater than a specific value.

d. Convert all temperatures to Fahrenheit.

e. Print the days had temperatures above the average

```python
temps = pd.Series(
    [30, 32, 29, 35, 33, 31, 28],
    index=["Monday", "Tuesday", "Wednesday", "Thursday",
           "Friday", "Saturday", "Sunday"]
)
```

```python
print("The temperatures are : \n")
temps
```

```python
print("The average of all temperatures is : ", temps.mean())
print("\nThe maximum temperature and its day is : ")
print("\nDay:", temps.idxmax())
print("\nTemperature:", temps.max())
print("\nThe minimum temperature and its day is : ")
print("\nDay:", temps.idxmin())
print("\nTemperature:", temps.min())
```

```python
value = int(input("Enter a specific temperature value : "))
print(f"Temperatures above the given value of {value} are : ")
print(temps[temps>value])
```

```python
fahrenheit = (temps*9/5) +32
print("The temperatures in fahrenheits are : ")
fahrenheit
```

```python
print(f"The temperatures above the mean temperature of {temps.mean():.2f} are : ")
temps[temps>temps.mean()]
```

Create a data frame with details of 10 students and columns as Roll Number, Name, Gender, Marks1, Marks2, Marks3.

a. Create a new column with total marks

b. Find the lowest marks in Marks1

c. Find the Highest marks in Marks2

d. Find the average marks in Marks3

e. Find student name with highest average

f. Find how many students failed in Marks2 (<40)

```python
students = pd.DataFrame({
    "Roll Number": [101,102,103,104,105,106,107,108,109,110],
    "Name": ["Arun","Rahul","Priya","Sneha","Rohan","Amit","Neha","Kiran","Anjali","Vikram"],
    "Gender": ["M","M","F","F","M","M","F","M","F","M"],
    "Marks1": [78,45,89,67,90,34,76,55,81,69],
    "Marks2": [88,39,76,91,52,45,67,33,84,71],
    "Marks3": [80,60,92,75,88,50,79,65,90,73]
})

students
```

```python
students["Total"] = students["Marks1"]+students["Marks2"]+students["Marks3"]
students
```

```python
print("Lowest marks in Marks1 : ", students["Marks1"].min())
print("Highest marks in Marks2 : ", students["Marks2"].max())
print("Average marks in Marks3 : ", students["Marks3"].mean())
```

```python
students["Average"] = students["Total"]/3
students
```

```python
print("Student with highest average:", students.loc[students["Average"].idxmax(), "Name"])
```

```python
failed = students["Marks2"]<40
print("Total number of people who failed in marks2 is : ", failed.sum())
```
