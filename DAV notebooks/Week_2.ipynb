1. Write a Python code block that inputs numbers into a list. Print the largest, smallest, the sum, and the average of the numbers.
Count occurrences of a specific number in the list.

```python
import numpy as np
numbers = []    # Initialising a blank list to get numbers appended into

a = int(input("Enter the number of elements you want to add : "))
for i in range(a):
    x = int(input("Enter a number : "))
    numbers.append(x)    # .append can be used to append a number to the list

print("The list of numbers is : ", numbers)
print("The largest number in list is : ", max(numbers))   # functions like max, min, sum can be used on basic lists for computation
print("The smallest number in list is : ", min(numbers))
print("The sum of all numbers in list is : ", sum(numbers))
print("The average of all numbers in list is : ", np.average(numbers))   # np.average can be used to find the average of all numbers in the list
```
